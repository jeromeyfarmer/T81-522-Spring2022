There are some spikes in flight delays in certain dates. Could they be associated with federal holidays? Let's create a new feature for each flight that captures the difference in days between flight days and the closest federal holidays.

### Creating new temporal features

```python
from datetime import date

holiday= [date(2015,1,1),date(2015,1,19),date(2015,2,16),  date(2015,5,25), date(2015,7, 3), date(2015,9,7), 
        date(2015,10,12), date(2015,11,11),date(2015,11,26),  date(2015,12,25)]

holiday = [pd.to_datetime(x) for x in holiday]

def dis_from_the_nearest_holiday(date, holiday):
    """ returns difference in days between a date and list of days in holiday"""
    
    return min([abs(date - x) for x in holiday])
    
```

```python
df['time_del'] = df['DATE'].apply(lambda x: dis_from_the_nearest_holiday (x, holiday))
```

```python
df['time_del'] = df['time_del'].dt.days
```

Now lets plot a scatterplot between `time_del` and `DEPARTURE_DELAY`.

```python
sns.scatterplot(x="time_del", y="DEPARTURE_DELAY", data=df)
```

It doesn't show the relationship between the two if any. We could also hypothesize that the delay might have been caused by certain weather events such as winter storms in Chicago. To test that hypothesis, we need to pull weather data. But for the sake of this exercise, we will just analyze if there are more delays in winter months. 

```python
df.groupby('MONTH')['DEPARTURE_DELAY'].mean()
```

Furthermore, we may want to create a different column, which captures  hourly information of the flight departure. The `DEPARTURE_TIME` column is represented as float with digits preceeding decimal indicating the hour time. 

```python
def extract_hour(x):
    
    try:
        return datetime.time(int(np.round(x)))
    except ValueError:
        return np.nan

df['HOUR'] = df["DEPARTURE_TIME"].apply(extract_hour)
```

```python
df.groupby('HOUR')['DEPARTURE_DELAY'].mean()
```
