 There are several vizualization packages in Python, some of which we will go through in the upcoming weeks. In this lesson, we will get familiar with basic vizualization using `.plot()` method of Pandas dataframe and series for quick data exploration. The plot method in Pandas is just a simple wrapper around `plot` method of `matplotlib.pyplot` module. 

Matplotlib is the foundational Python plotting library. It lets you control all aspects of a graphical plot, however, it also requires you to write verbose codes to get reasonably good looking plots. The plot method in Pandas lets you use some of the matplotlib's plotting capability with less amount of codes for the same results.

To explore some of the vizualization capability, lets use the same beer dataset that we used in the previous exercise.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
INPUT_FILE = "/home/asimbanskota/t81_577_data_science/weekly_materials/week6/data/beer_reviews.csv"
df = pd.read_csv(INPUT_FILE)
```

```python
from datetime import datetime
def compute_year(x):
    """ returns the year info from unix timestap"""
    return datetime.fromtimestamp(x).year

df['review_year']= df.review_time.apply(compute_year)
```

```python
df['review_year'].hist();
```

```python
df['review_year'].hist(bins=12, facecolor='yellow', edgecolor='gray');
```

To create a scatter plot in Pandas we can call dataframe.plot.scatter() and pass it the name of the x-column and y-column as arguments. Optionally we can also pass it a title.

```python
dfs = df.sample(frac = 0.1)

dfs.plot.scatter(x= 'review_taste', y = 'review_palate', title = 'Scatter plot');
```

```python
dfs.plot.hexbin(x= 'review_taste', y = 'review_palate', gridsize=10)
```
